### __3 PLOTS IN 1 FIGURE__

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,5,6,5,7,8,7]
xy = pd.DataFrame({'X':x, 'Y':y})

plt.plot(xy['X'], xy['Y'])
plt.plot(xy['Y'], xy['X'])
plt.plot(xy['X'], xy['X'])
plt.legend(['X and Y', 'Y and X', 'X and X'], loc=4)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,5,6,5,7,8,7]
xy = pd.DataFrame({'X':x, 'Y':y})

plt.plot(xy['X'], xy['Y'], xy['Y'], xy['X'], xy['X'], xy['X'])
```

```python
plt.plot(xy['X'], xy['X'])
plt.plot(xy['X'], xy['X'] ** 2)
plt.plot(xy['X'], xy['X'] ** 3)
plt.legend(['X and X', 'X and X^2', 'X and X^3'], loc=0)
plt.show()
```

### __STYLING PLOT LINE__
##### _color / c = warna garis ( r=red, g=green, b=blue, c=cyan, m=magenta)_
##### _linewidth / lw = tebal garis_
##### _linestyle / ls = tipe garis (bisa pakai --, :, dst)_
##### _marker = data plotnya dikasih mark (bisa pakai o, D(diamond), * dst)_
##### _markersize / ms = ukuran markernya_
##### _mec = warna border dari markernya_
##### _mfc = warna markernya (by default warna marker ikutin color / warna garis)_

```python
plt.plot(xy['X'], xy['X'], color='red', linewidth=2)
plt.plot(xy['X'], xy['X'] ** 2, c='yellow', lw=1, linestyle='--')
plt.plot(xy['X'], xy['X'] ** 3, ls=':', marker='$Halo$', markersize=30, mec='red', mfc='yellow')
plt.legend(['X and X', 'X and X^2', 'X and X^3'], loc=0)
plt.show()
```

### __STYLING GRID LINE__

```python
plt.plot(xy['X'], xy['X'])
plt.plot(xy['X'], xy['X'] ** 2)
plt.plot(xy['X'], xy['X'] ** 3)
plt.legend(['X and X', 'X and X^2', 'X and X^3'], loc=0)
plt.grid(True, color='r', ls='--', lw=0.5)
plt.show()
```

### __STYLING SHORTCUT__

```python
plt.plot(
    xy['X'], xy['X'], 'or-',
    xy['X'], xy['X'] ** 2, 'g--D',
    xy['X'], xy['X'] ** 3, 'b:*')
plt.show()
```

### __STYLING TICKS__

### __ADJUSTING PLOT__
##### _adjust pertama di visual code untuk menentukan value left bottom right top_

```python
x = np.arange(0, 10)
y = [2,1,3,5,6,4,7,9,8,9]

plt.plot(x, y)
plt.subplots_adjust(left=0.09, bottom=0.18, right=0.94, top=0.9)
plt.show()
```

### __TEXT CUSTOM__
##### _plt.text(titik x, titik y, textnya)_
##### _plt.annotate(textnya, titik yg ditunjuk, titik text, arrowprops = {propertinya}_

```python
plt.plot(xy['X'], xy['Y'],c='black', marker='o', mfc='y')
plt.fill_between(xy['X'], xy['Y'], facecolor='g', alpha=0.6)
plt.text(2,4, 'HAI',fontdict={'c':'blue'})
plt.show()
```

```python
plt.plot(xy['X'], xy['Y'])
for i in xy['X']:
    plt.text(i - .2, xy['Y'].iloc[i-1] + .15, '%d,%d' %(i,xy['Y'].iloc[i-1]))

plt.annotate(
    'Peak Point', xy=(8,8), xytext = (4,8),
    arrowprops = dict(facecolor='y'))
plt.yticks(np.arange(0, 11, 2))
plt.show()
```
