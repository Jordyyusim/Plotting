```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = pd.DataFrame({
    'Massa' : np.arange(40, 100, 5),
    'Tinggi' : np.arange(160, 172, 1)
})
x.head(3)
```

## _Creating Subplots from one figure_

### Cara 1
#### Using Subplot

```python
plt.figure('Massa & Tinggi', figsize=(12,6))

plt.subplot(121)
plt.plot(x['Massa'], x['Tinggi'])

plt.subplot(122)
plt.plot(x['Tinggi'], x['Massa'])
```

### Cara 2
#### Using add_subplot but make a figure variable first 

```python
fig = plt.figure('Massa & Tinggi', figsize=(12,6))

fig.add_subplot(121)
plt.plot(x['Massa'], x['Tinggi'])

fig.add_subplot(122)
plt.plot(x['Tinggi'], x['Massa'])
```

### Cara 3
#### Using subplots(total row, total col)

```python
fig, [ax1, ax2] = plt.subplots(1, 2)
ax1.plot(x['Massa'], x['Tinggi'])
ax2.plot(x['Tinggi'], x['Massa'])
plt.show()
```

### Cara 4
#### Using subplot2grid((shape), (location))
#### shape(row, col)
#### location(row, col) starts from 0

```python
ax1 = plt.subplot2grid((2,2), (0,0))
ax2 = plt.subplot2grid((2,2), (0,1))
ax3 = plt.subplot2grid((2,2), (1,0))
ax4 = plt.subplot2grid((2,2), (1,1))
ax1.plot(x['Massa'], x['Tinggi'])
ax2.plot(x['Massa'], x['Tinggi'], 'r--o')
ax3.plot(x['Massa'], x['Tinggi'], 'gD:')
ax4.plot(x['Massa'], x['Tinggi'], 'y*')
plt.show()
```
