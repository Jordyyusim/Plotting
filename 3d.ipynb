```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
fig = plt.figure()
axes = plt.subplot(111, projection='3d')
x = np.arange(1, 11)
y = np.arange(1, 11)
z = np.arange(1, 11) #harus 2 dimensi
axes.plot_wireframe(x, y, np.array([z]), color='r', linestyle='--')

axes.set_xlabel('Nilai X')
axes.set_ylabel('Nilai Y')
axes.set_zlabel('Nilai Z')
plt.show
```

```python
fig = plt.figure()
axes = plt.subplot(111, projection='3d')
x = np.arange(1, 11)
y = np.arange(1, 11)
z = np.arange(1, 11) #harus 2 dimensi
axes.scatter(x, y, np.array([z]), color='r', marker="*", s=100)

axes.set_xlabel('Nilai X')
axes.set_ylabel('Nilai Y')
axes.set_zlabel('Nilai Z')
plt.show

# Warna transparan artinya jauh
# Warna terang artinya dekat
```

```python
fig = plt.figure()
axes = plt.subplot(111, projection='3d')
x = np.arange(1, 11); y = np.arange(1, 11); z = np.zeros(10)

lx = np.ones(10)
ly = np.ones(10)
tz = np.arange(1, 11)
axes.bar3d(x, y, z, lx, ly, tz,
          color=['r', 'g', 'b', 'c', 'm'] * 2)

axes.set_title(label='3D BAR')
axes.set_xlabel('Nilai X')
axes.set_ylabel('Nilai Y')
axes.set_zlabel('Nilai Z')
plt.show
```

```python
fig = plt.figure()
axes = plt.subplot(111, projection='3d')
x = np.arange(1, 11); y = np.arange(1, 11); z = np.arange(1, 11)

lx = np.full((10,), 3)
ly = np.ones(10)
tz = np.arange(1, 11)
# tz = np.ones(10)
axes.bar3d(x, y, z, lx, ly, tz)

axes.set_xlabel('Nilai X')
axes.set_ylabel('Nilai Y')
axes.set_zlabel('Nilai Z')
plt.show
```
